```python
import requests
import bs4
import pandas as pd
```

```python
url = 'https://www.intronesia.com/intro/dari-34-provinsi-di-indonesia-jumlah-kabupaten-dan-kota-di-indonesia-adalah-416-kabupaten-dan-98-kota'
res = requests.get(url)
soup = bs4.BeautifulSoup(res.text, "html.parser")
print(soup.title.string)
```

```python
provinceTable = soup.table
provinceTable
```

```python
provinceTable.findAll("tr")[0].findAll('th')
```

```python
t1 = [] 
t2 = [] 
t3 = [] 
t4 = []
states = []

for item in provinceTable.findAll("tr"):
    if(len(states) == 0):
        states = item.findAll('th')
    cells = item.findAll('td')
    if(len(cells) == 4):
        t1.append(cells[0].find(text=True))
        t2.append(cells[1].find(text=True))
        t3.append(cells[2].find(text=True))
        t4.append(cells[3].find(text=True))
```

```python
t1
```

```python
df = pd.DataFrame()
df["Provinsi"] = t1
df["Kabupaten"] = t2
df["Kota"] = t3
df["Total"] = t4
```
